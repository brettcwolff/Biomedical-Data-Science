## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load dataset

```python
data = pd.read_csv('HW3Data.csv', header = None)
data
```

## Access rows/columns of labels/features

```python
L = data.loc[:,0].values
F = data.loc[:,1:].values
```

## Import library

```python
from sklearn.model_selection import train_test_split
```

## Split data into train and test subsets 

```python
X_train, X_test, y_train, y_test = \
    train_test_split(F, L,
                    test_size = 0.2,
                    stratify = L,
                    random_state = 1)
```

## Import libraries

```python
from sklearn.svm import SVC
from sklearn.feature_selection import RFECV
```

## Recursive feature elimination with cross-validation

```python
svc = SVC(kernel = "linear")
rfecv = RFECV(estimator = svc, step = 1, cv = 5,scoring = 'accuracy')
rfecv.fit(X_train, y_train)

print("Optimal number of features : %d" % rfecv.n_features_)
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score(nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

## Reporting the performance 

```python
svc.fit(X_train[:,rfecv.support_], y_train)
svc.score(X_test[:,rfecv.support_], y_test)
```
